```python
import geopandas as gpd
from mapreader import SheetDownloader
```

```python
parks = gpd.read_file("./data/National_Parks_England_4829975525583709119.geojson")
south_downs = parks[parks["NAME"] == "SOUTH DOWNS"]
```

```python
# 1949-72

my_ts_1949_1972 = SheetDownloader(
    metadata_path="./data/api.geojson",
    download_url="https://geo.nls.uk/maps/tiles/os/britain10knationalgridnew/{z}/{x}/{y}.png"
)

# Filter metadata up to 1972
my_ts_1949_1972.extract_published_dates()
my_ts_1949_1972.metadata = my_ts_1949_1972.metadata[my_ts_1949_1972.metadata["published_date"] <= 1972]

# Query and download
my_ts_1949_1972.get_grid_bb(17)
my_ts_1949_1972.query_map_sheets_by_polygon(south_downs.iloc[0].geometry, mode="intersects")
my_ts_1949_1972.download_map_sheets_by_queries(
    force=True, download_in_parallel=False, path_save="maps_south_downs_1949_1972"
)
```
